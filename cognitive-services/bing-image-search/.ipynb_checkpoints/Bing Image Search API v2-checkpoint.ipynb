```python
def split_search_term(search):
    # To Parse out the Macys.com site identifieer
    search = search.split(" ")
    search_term = ""
    for i in range(len(search)):
        #print(str(i))
        if i == len(search) -1:
            search_term += search[i]
        else:
            if i != 0:
                search_term += search[i] + " "
    return search_term.strip()
```
